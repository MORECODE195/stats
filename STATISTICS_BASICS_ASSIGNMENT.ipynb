# Q3.  Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?

 ANSWER. The term dispersion is a statistical term which represents the size of distribution of values that are expected for a specific variable. Dispersion can be calculated and measured using different statistics such as, range, variance and standard deviation. Smaller values for variance and standard deviation indicate that the data is more tightly grouped around the mean, while larger values indicate that the data is more spread out. If all the values in a data set are the same, then the variance and standard deviation are zero.

 ANSWER. Measure of Symmetrical is done using skewness. skewness is a measure of how symmetrical a distribution is. There are Three types of Skewness :- 1- Positive skew, tail is on the right side of distribution. 2- Negative skew, tail is on the left side of the distribution. 3- Zero skew, when the distribution is symmetrical and the left and right sides are mirror images. # Q7. What is the interquartile range (IQR), and how is it used to detect outliers?

 ANSWER. The Interquartile range (IQR) is a measure of statistical dispersion which is the spread of the data. It is defined as the difference between the 75th and 25th percentiles of the data. It is used to detect the outliers in the data set by calculating the lower and upper bounds of the data.

# Q9.  Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).

 ANSWER. The empirical rule states that in a normal distribution, virtually all observed data will fall within three standard deviations of the mean. Under this rule, 68% of the data will fall within one standard deviation, 95% within two standard deviations, and 99.7% within three standard deviations from the mean.
